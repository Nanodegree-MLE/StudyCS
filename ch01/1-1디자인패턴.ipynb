# 1.1 디자인패턴

- 프로그램을 설계할 때 발생했던 문제점들을 객체 간의 상호 관계 등을 이용해서 해결할 수 있도록 하나의 '규약' 형태로 만들어 놓은 것을 의미
- '반복적으로 일어나는 문제들을 어떻게 풀어나갈 것 인가'에 대한 일종의 솔루션

## 왜 디자인 패턴이 필요할까?
- 설계자로 하여금 재사용이 가능한 설계는 선택하고, 재사용을 방해하는 설계는 배제하도록 도와준다.
- 이미 만든 시스템의 유지보수나 문서화도 개선,클래스의 명세도 정확하게, 객체 간의 상호작용 또는 설계 의도까지 명확하게
<br>

결론적으로, **최소한의 비용**으로 **재사용성, 유지보수, 효율적, 생산적** 위함!

## 클래스와 객체

`클래스 객체명 = new 클래스();`<br>
`붕어빵틀 붕어빵 = new 붕어빵틀();`<br>
`금형기계 붕어빵틀 = new 금형기계();`<br>

새로운 금형기계를 만들었는데, 붕어빵틀이 되었다????????<br>

붕어빵에게 붕어빵 틀이란?

붕어빵을 만드는 팩토리였던 것이다.

1. 사람은 클래스인가? 객체인가?<br>
2. 김연아는 클래스인가? 객체인가?<br>
3. 뽀로로는 클래스인가? 객체인가?<br>
4. 펭권은 클래스인가? 객체인가?<br>

클래스와 객체를 구분하는 간단한 방법은 나이를 물어보는 것이다.

1. 사람의 나이는 몇 살인가?
2. 김연아의 나이는 몇 살인가?
3. 뽀로로의 나이는 몇 살인가?
4. 펭귄의 나이는 몇 살인가?

사람과 펭귄의 나이는 대답할 수 없다. 김연아와 뽀로로는 구글에 치면 언제 태어났는지 너무나도 잘 나온다.<br>
**클래스는 분류! 객체는 실체**

```
클래스 : 객체 = 사람 : 김연아 = 펭귄 : 뽀로로
```
