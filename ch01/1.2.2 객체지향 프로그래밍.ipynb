```java
// λ„ν•μ„ λ‚νƒ€λ‚΄λ” κΈ°λ³Έ μ¶”μƒ ν΄λμ¤
abstract class Shape {
    abstract double calculateArea();
}

// μ›μ„ λ‚νƒ€λ‚΄λ” ν΄λμ¤
class Circle extends Shape {
    private double radius;

    public Circle(double radius) {
        this.radius = radius;
    }

    public double getRadius() {
        return radius;
    }

    public void setRadius(double radius) {
        this.radius = radius;
    }

    @Override
    double calculateArea() {
        return Math.PI * radius * radius;
    }
}

// μ‚¬κ°ν•μ„ λ‚νƒ€λ‚΄λ” ν΄λμ¤
class Rectangle extends Shape {
    private double width;
    private double height;

    public Rectangle(double width, double height) {
        this.width = width;
        this.height = height;
    }

    public double getWidth() {
        return width;
    }

    public void setWidth(double width) {
        this.width = width;
    }

    public double getHeight() {
        return height;
    }

    public void setHeight(double height) {
        this.height = height;
    }

    @Override
    double calculateArea() {
        return width * height;
    }
}

// λ„ν• κ³„μ‚°μ„ λ‹΄λ‹Ήν•λ” ν΄λμ¤
class AreaCalculator {
    public static double calculateArea(Shape shape) { 
		// abstract class Shape μ calculateArea μ™€ μ΄λ¦„μ΄ κΌ­ κ°™μ§€λ” μ•μ•„λ„ λλ‹¤.
        return shape.calculateArea();
    }
}

// μ‚¬μ© μμ‹
public class Main {
    public static void main(String[] args) {
        Circle circle = new Circle(5.0);
        double circleArea = AreaCalculator.calculateArea(circle);
        System.out.println("μ›μ λ„“μ΄: " + circleArea);

        Rectangle rectangle = new Rectangle(4.0, 6.0);
        double rectangleArea = AreaCalculator.calculateArea(rectangle);
        System.out.println("μ‚¬κ°ν•μ λ„“μ΄: " + rectangleArea);
    }
}
```
