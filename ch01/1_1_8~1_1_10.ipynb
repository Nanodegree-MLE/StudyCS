### MVC 패턴
모델(Model), 뷰(View), 컨트롤러(Controller)로 이루어진 디자인 패턴

IMG_213EF1323ADF-1.jpeg
Fig 1. MVC 패턴


## 모델
애플리케이션의 데이터인 데이터베이스, 상수, 변수 등을 의미

**애플리케이션** : 특정한 목적을 위해 설계되고 개발된 소프트웨어

e.g. 사각형 모양의 박스 안에 글자가 들어있을 때, 박스 위치 정보, 글자 내용, 글자 위치 등 관련 정보

## 뷰
사용자의 인터페이스(사용자가 볼 수 있는 화면)

=> 사용자에게 정보를 표시하고, 사용자 입력을 받아 컨트롤러에 전달(변경이 일어나면 컨트롤러에 이를 전달해야 함)

## 컨트롤러
사용자의 입력을 받아 모델에게 전달하고, 모델의 상태 변경에 따라 뷰를 업데이트함(메인 로직을 담당). 즉, 사용자 입력을 처리하고 그에 따른 모델의 변경을 관리하며, 이를 뷰에 반영.

## 예시

**MVC 패턴 예시: 온라인 쇼핑몰**<br/><br/>


모델: 제품 정보, 주문 정보, 사용자 정보 등 데이터를 저장하고 관리 & 주문 처리, 결제 처리 등 비즈니스 로직을 담당<br/><br/>

뷰: 제품 목록 페이지, 제품 상세 페이지, 장바구니 페이지, 결제 페이지 등 사용자 인터페이스를 구성<br/><br/>

컨트롤러: 사용자 입력을 받아 모델에 요청을 전달하고 모델에서 변경된 데이터를 받아 뷰에 반영 <br/><br/>

e.g. 사용자가 제품을 검색 시, 컨트롤러는 모델에 검색 요청을 전달하고 모델에서 받은 검색 결과를 뷰에 표시

## MVC 패턴의 특징
애플리케이션의 구성 요소를 세 가지 역할로 구분하여 개발 프로세스에서 각각의 구성 요소에만 집중해서 개발할 수 있음<br/><br/>

**장점** : 재사용성과 확장성이 용이함

=> 모델, 뷰, 컨트롤러 각각의 구성 요소는 독립적으로 작동하기 때문에 다른 프로젝트에서도 쉽게 재사용할 수 있음.

e.g. 이미 개발된 제품 모델을 다른 쇼핑몰 프로젝트에서도 활용할 수 있습니다.<br/><br/>


**단점** : 애플리케이션이 복잡해질수록 모델과 뷰의 관계가 복잡해짐

# MVP 패턴
MVC 패턴으로부터 파생되었으며, MVC에서 C에 해당하는 컨트롤러가 프레젠터(presenter)로 교체된 패턴

=>  UI(View)와 비즈니스 로직(Model)을 더욱 분리하고 상호작용을 간소화하는 데 중점을 둠

=> 뷰와 프레젠터는 일대일 관계로, MVC 패턴보다 더 강한 결합을 지님(책)

# MVVM 패턴
MVC의 C에 해당하는 컨트롤러가 view model로 바뀐 패턴

**view model** : view를 더 추상화한 계층

=> UI를 별도의 코드 수정 없이 재사용할 수 있음
