### 1.1.4. 옵저버 패턴<br>
옵저버(관찰자)들이 관찰하고 있는 대상자의 상태가 변화가 있을 때마다 대상자는 직접 목록의 각 관찰자들에게 통지하고, 관찰자들은 알림을 받아 조치를 취하는 행동 패턴이다.<br><br>
<img src="/images/observer_pattern.jpg" width="450px" height="300px"></img><br/>

이 패턴을 이해하는데 있어 뉴스 피드나 유튜브로 비유해보면 쉽다. 유튜브 채널은 주체(Subject)가 되고 구독자들은 관찰자(Observer)가 되는 구조로 보면 된다. <br><br>
실제로 유튜버가 영상을 올리면 여러명의 구독자들은 모두 영상이 올라왔다는 알림을 받는데, 이를 패턴 구조로 보자면 구독자들은 해당 채널을 구독함으로써 채널에 어떠한 변화(영상을 올리거나 커뮤니티에 글을 쓰거나)가 생기게 되면 바로 연락을 받아 탐지하는 것이다.<br> <br>
반면 구독을 해지하거나 안한 시청자에게는 알림이 가지않게 된다.<br><br>

##### 옵저버 패턴 사용 시기 <br>


* 앱이 한정된 시간, 특정한 케이스에만 다른 객체를 관찰해야 하는 경우 대상 객체의 상태가 변경될 때마다 다른 객체의 동작을 트리거해야 할때.<br><br>
* 한 객체의 상태가 변경되면 다른 객체도 변경해야 할때. 그런데 어떤 객체들이 변경되어야 하는지 몰라도 될 때.<br><br>
* MVC 패턴에서도 사용됨 (Model, View, Controller)<br><br>
    - MVC의 Model과 View의 관계는 Observer 패턴의 Subject 역할과 Observer 역할의 관계에 대응된다.
* 하나의 Model에 복수의 View가 대응한다.

 
##### 패턴 장점<br>

* Subject의 상태 변경을 주기적으로 조회하지 않고 자동으로 감지할 수 있다.<br><br>
* 발행자의 코드를 변경하지 않고도 새 구독자 클래스를 도입할 수 있어 개방 폐쇄 원칙(OCP)Visit Website 준수한다.<br><br>
* 런타임 시점에서에 발행자와 구독 알림 관계를 맺을 수 있다.<br><br>
* 상태를 변경하는 객체(Subject)와 변경을 감지하는 객체(Observer)의 관계를 느슨하게 유지할 수 있다. (느슨한 결합)<br><br>

 
##### 패턴 단점<br>
* 구독자는 알림 순서를 제어할수 없고, 무작위 순서로 알림을 받음 <br><br>
    - 하드 코딩으로 구현할수는 있겠지만, 복잡성과 결합성만 높아지기 때문에 추천되지는 않는 방법이다. <br><br>
* 옵저버 패턴을 자주 구성하면 구조와 동작을 알아보기 힘들어져 코드 복잡도가 증가한다. <br><br>
* 다수의 옵저버 객체를 등록 이후 해지하지 않는다면 메모리 누수가 발생할 수도 있다. <br><br>

#### 주체와 객체의 분리 및 결합
옵저버 패턴에서 주체(관찰 가능 항목이라고도 함)와 옵저버의 역할은 서로 다르지만 보완적입니다.<br><br> 
이 패턴에 대한 표준 접근 방식은 객체와 주체를 결합해야 하는지 분리해야 하는지를 엄격하게 규정하지 않습니다. <br><br>
대신 이들 간의 관계에 중점을 두며, 그 관계로는 종속성(dependancy)과 알림(notification)이 있습니다. <br><br>
구현의 유연성을 통해 패턴을 애플리케이션의 특정 요구 사항에 맞게 조정할 수 있습니다.<br><br>

##### 분리
일반적으로 관찰자 패턴은 주체와 객체를 명확하게 구분합니다.

주체는 관찰자 목록을 유지하고 상태 변화를 알립니다.<br><br>
옵저버는 주체를 구독하여 업데이트를 받고 이러한 업데이트에 반응하는 방법을 정의합니다.<br><br>
이러한 분리는 관심사 분리 원칙을 준수하여 모듈성을 강화하고 유지 관리 및 테스트를 용이하게 합니다.<br><br>
주체는 옵저버가 업데이트로 수행하는 작업의 세부 사항에 관심이 없습니다. 그들은 변경 사항을 알리는 데 중점을 둡니다. <br><br>
반면에 옵저버는 주체가 자신의 상태 또는 관찰자 목록을 관리하는 방법을 이해할 필요 없이 변경 사항에 대응하는 방법을 자유롭게 결정할 수 있습니다.<br><br>

##### 결합
객체와 주체의 역할을 결합하는 것이 실용적일 수 있는 시나리오가 있습니다.<br><br>
* 작거나 간단한 애플리케이션: 복잡도가 낮은 애플리케이션에서는 피사체와 관찰자를 결합하면 코드가 더 단순해질 수 있습니다. 그러나 이 접근 방식은 유연성을 감소시키고 시스템을 확장하거나 수정하기 어렵게 만들 수 있습니다.<br><br>
* 성능 최적화: 성능이 중요한 애플리케이션에서는 개체 수를 줄이고 상호 작용을 단순화해야 할 수도 있습니다. 비록 모듈성이 감소하더라도 역할을 결합하는 것이 이를 달성하는 방법이 될 수 있습니다.<br><br>

#### 상속과 구현

##### 상속
정의: 상속을 통해 클래스(자식 클래스)가 다른 클래스(상위 클래스)로부터 특성과 메서드를 상속할 수 있습니다. 이 메커니즘은 코드를 구성하고 재사용하는 방법을 제공합니다. <br><br>
특성<br><br>
코드 재사용: 하위 클래스는 다시 작성할 필요 없이 상위 클래스의 코드를 재사용합니다.<br><br>
계층 구조: 상위 클래스와 하위 클래스 사이에 명확한 계층 관계를 설정합니다.<br><br>
재정의: 하위 클래스는 상위 클래스에서 상속된 메서드를 재정의하여 특수한 동작을 제공할 수 있습니다.<br><br>

용도<br><br>
상속은 일반적으로 하위 클래스와 상위 클래스 사이에 자연스러운 "is-a" 관계가 있을 때 사용됩니다. 예를 들어, 개는 동물이기 때문에 'Dog' 클래스는 'Animal' 클래스에서 상속받을 수 있습니다.<br><br>

##### 구현(인터페이스 및 추상 클래스)
정의: Python에는 Java와 같은 언어와 같은 방식의 인터페이스가 없습니다. 대신, Python의 덕 타이핑 철학과 ABC(추상 기본 클래스)를 사용하여 클래스가 특정 메서드를 구현하는지 확인합니다. <br><br>
특성 <br><br>
* 계약 시행: ABC는 파생 클래스가 특정 메서드를 구현하도록 강제할 수 있습니다. 이는 다른 언어의 인터페이스와 유사합니다.<br><br>
* 코드 상속 없음: ABC는 기본 구현을 제공할 수 있지만 기능 상속보다는 특정 메서드가 구현되도록 하는 데 중점을 둡니다.<br><br>
* 다중 구현: 단일 클래스는 필요한 메서드를 제공하여 단일 상속에 비해 더 많은 유연성을 제공함으로써 여러 인터페이스(또는 추상 기본 클래스)를 구현할 수 있습니다.<br><br>

용도: 이 접근 방식은 클래스가 특정 계약이나 API를 충족하는지 확인해야 할 때 사용됩니다. 계층적 관계보다는 클래스가 수행할 수 있는 기능(할 수 있음) 또는 동작에 관한 것입니다.<br><br>
##### 차이점 <br><br>
관계 대 계약: 상속은 계층적 관계를 암시하는 "is-a" 관계를 나타냅니다. 특히 Python의 추상 기본 클래스를 통한 구현은 클래스가 특정 동작을 수행할 수 있음을 나타내는 계약 이행에 중점을 둡니다.<br><br>
코드 재사용: 상속은 주로 코드 재사용 및 기능 확장에 사용됩니다. 구현(ABC 사용)은 클래스가 코드 재사용에 덜 중점을 두고 특정 메서드를 제공하는지 확인하는 데 사용됩니다.<br><br>
유연성: Python은 다중 상속을 허용하지만 복잡하고 디버그하기 어려운 문제가 발생할 수 있습니다. 여러 추상 기본 클래스 또는 인터페이스를 구현하면 다중 상속의 복잡성 없이 다양한 동작이나 기능을 결합하는 더 깔끔한 방법을 제공할 수 있습니다.<br><br>
